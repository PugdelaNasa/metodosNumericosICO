### Precision y Exactitud

**Exactitud (Cercania al valor verdadero)**
Mide que tan cerca esta un resultado de su valor verdadero o exacto.
Ejemplo: si la raíz exacta de una ecuación es 2.0000 y un método entrega 2.0001, el resultado es muy exacto porque está cerca del valor verdadero.


**Precision (Consistencia o nivel de detalle de los resultados)**
Mide el grado de detalle o consistencia en los resultados obtenidos, es decir, cuántas cifras significativas se conservan o cuán repetibles son las aproximaciones.
Un resultado puede ser muy preciso (muchas cifras decimales calculadas) pero poco exacto (si está lejos del valor verdadero).
Ejemplo: si un método entrega 3.141592 con varias cifras decimales, es preciso, pero si el valor verdadero era 2.718281, no es exacto.

### Errores en Calculos Numericos
A menudo en los algoritmos, nos encontramos con errores de redondeo, truncamiento, etc., que pueden afectar el resultado final.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
 #Error en valores aproximados
valor_exacto = 1/3
valor_aproximado = 0.3333333333

error_absoluto = abs (valor_exacto - valor_aproximado)
print("Error absoluto", error_absoluto)
```

```python
# Error en redondeo
numero_decimal = 0.123456789

# Redondear a 4 cifras decimales
numero_redondeado = round(numero_decimal, 4)

# Calcular el error de redondeo
error_redondeo = abs(numero_decimal - numero_redondeado)

print("Número original:    ", numero_decimal)
print("Número redondeado:  ", numero_redondeado)
print("Error de redondeo: ", error_redondeo)  
```

```python
import math

e_aprox = sum(1/math.factorial(n) for n in range(0,10))
e_exact = math.e

error_truncamiento = abs(e_exact - e_aprox)

print("Error de truncamiento: ", error_truncamiento)
```

### Estabilidad y Convergencia
De igual forma, en los métodos numéricos debemos tener en cuenta la estabilidad y convergencia de los algoritmos.

```python
# Estabilidad
suma = 0.0

for i in range(10000):
    suma += 0.1

print("Suma Acumulada: ", suma)
```

```python
# Convergencia
def funcion(x):
    return (2 * x + 3) / 5

def metodo_iterativo(x0, tolerancia=1e-6, max_inter=1000):
    x = x0
    for i in range(max_inter):
        x1 = funcion(x)
        if abs(x1 - x) < tolerancia:
            return x1, i
        x = x1
    return None, i

resultado, iteraciones = metodo_iterativo(0.0)
print("Resultado: ", resultado, "Iteraciones: ", iteraciones)
```
